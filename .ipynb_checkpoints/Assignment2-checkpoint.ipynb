```python
# Dependencies
import pandas as pd
import os
```

```python
# Load in File from resources
# 'movie_scores.csv'
csvfile = os.path.join("Resources","movie_scores.csv")
fileoutput = os.path.join("lesser_movies.csv")
```

```python
# Read and display with pandas
file_one_df = pd.read_csv(csvfile, encoding="UTF-8")

file_one_df.head()
```

```python
# List all the columns the table provides
# .columns will also get you this
file_one_df.nunique()
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relatingto IMDB
# Could just pick the columns you need instead
film_df = file_one_df.set_index("FILM")
IMDB_df = film_df.loc[:,["IMDB", "IMDB_norm","IMDB_norm_round","IMDB_user_vote_count"]]
IMDB_df.head()
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
# over_seven_df = IMDB_df["IMDB"]> 7
only_over_seven_df = IMDB_df.loc[IMDB_df["IMDB"] > 7,:]
only_over_seven_df.head()
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
less_watched_df = only_over_seven_df.loc[only_over_seven_df["IMDB_user_vote_count"] < 20000,:]
less_watched_df.head()
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
less_watched_df.to_csv(fileoutput, index=False, header=True)
less_watched_df.head ()
```
