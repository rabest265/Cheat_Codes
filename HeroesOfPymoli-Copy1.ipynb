# Heroes Of Pymoli Data Analysis

## Trend Summary
* Even through the males make up a much larger percentage of the players (84%), the female players spend more on average per player (4.47 compared to 4.07) and the Other/undisclosed catergory spent the most per person (4.57).

* Almost half the players fall in the 20-24 age range (44.8%), with the next most populous age ranges at 15-19 (18.60%) and 25-29 (13.4%).

* Of the 183 unique items, only one was bought more than 10 times ("Oathbreaker, Last Hope of the Breaking Storm" with 12.)	

* Of the 576 players only two spend more than 15 dollars (Lisosia93 with 18.96, and Idastidru52 with 15.45)

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt  # Used for charting


# File to Load
file_to_load = os.path.join("Resources","purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head ()
```

```python
# Display the total number of players
totplay = purchase_data["SN"].nunique()
print ("The total number of players is: " + str(totplay))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#Run basic calculations to obtain number of unique items, average price, etc.
totItem = purchase_data["Item ID"].nunique()
avgprice = purchase_data["Price"].mean()
totpurch = purchase_data["Item ID"].count()
totrev = purchase_data["Price"].sum()

# Create a summary data frame to hold the results
purchase_summary_df = pd.DataFrame({"Number of Unique Items": [totItem],
                          "Average Price": [avgprice],
                          "Number of Purchases": [totpurch],
                          "Total Revenue": [totrev]
                          })

# Optional: give the displayed data cleaner formatting
purchase_summary_df["Average Price"] = purchase_summary_df["Average Price"].map("${:,.2f}".format)
purchase_summary_df["Total Revenue"] = purchase_summary_df["Total Revenue"].map("${:,.2f}".format)

# Display the summary data frame
purchase_summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
# Count only the unique values of Gender per SN
gender_df = purchase_data.groupby('Gender').nunique()[["SN"]]

#Calculate the percentage field
gender_df["Percentage of Players"] = round(gender_df/totplay*100,2)
#Rename columns
gender_df = gender_df.rename(columns={"SN":"Total Count"})
#Sort columns by counts
gender_df = gender_df.sort_values ("Total Count", ascending=False)

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#Create a grouped dataframe
gender_purch_df = purchase_data.groupby('Gender')

#Create the summary table and add elements from the grouped dataframe
gender_purch_summary = purchase_data.groupby('Gender').count()[["SN"]]
gender_purch_summary = gender_purch_summary.rename(columns={"SN":"Purchase Count"})
gender_purch_summary["Average Purchase Price"] = gender_purch_df.mean()[["Price"]]
gender_purch_summary["Total Purchase Value"] = gender_purch_df.sum()[["Price"]]
gender_purch_summary["Avg Total Purchase per Person"] = gender_purch_summary["Total Purchase Value"]/gender_purch_df["SN"].nunique()

# Optional: give the displayed data cleaner formatting
gender_purch_summary["Average Purchase Price"] = gender_purch_summary["Average Purchase Price"].map("${:,.2f}".format)
gender_purch_summary["Total Purchase Value"] = gender_purch_summary["Total Purchase Value"].map("${:,.2f}".format)
gender_purch_summary["Avg Total Purchase per Person"] = gender_purch_summary["Avg Total Purchase per Person"].map("${:,.2f}".format)
gender_purch_summary = gender_purch_summary.rename(columns={"SN":"Purchase Count"})

# Display the summary data frame
gender_purch_summary
```

```python
# PIE CHARTS - Showing items as part of a whole
labels = gender_purch_summary.index.tolist()
sizes = gender_purch_summary["Purchase Count"]
colors = ["pink", "lightblue", "orange"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0.1, 0.1) # seperate each portion away from the other slices

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal") # forces pie charts to have a square border (no ovals)
plt.show()
```

## Age Demographics

* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
# Establish bins for ages
bins = [0,9,14,19,24,29,34,39,101]
group_names = ["<10","10-14","15-19", "20-24", "25-29","30-34","35-39", "40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names) 

# Calculate the numbers and percentages by age group
group_data = purchase_data.groupby("Age Range")
total = purchase_data["SN"].nunique()
count = group_data["SN"].nunique()
pct = count/total * 100

# Create a summary data frame to hold the results
summary_df = group_data.nunique()[["SN"]]
summary_df = summary_df.rename(columns={"SN":"Total Count"})
summary_df["Percentage of Players"] = pct

# Optional: round the percentage column to two decimal points
summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map("{:.2f}".format)

# Display Age Demographics Table
summary_df
```

```python
age_range = summary_df["Total Count"]
x_axis = np.arange(len(age_range))

plt.bar(x_axis, age_range, color='r', alpha=0.5, align="center", ) # center is where it is on the tick

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, group_names, rotation=45)# rotation is optional

plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(age_range)+50)

plt.show()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
#Create a grouped dataframe
SN_df = purchase_data.groupby("SN")

#Create the summary table and add elements from the grouped dataframe
SN_summary = SN_df.count()[["Purchase ID"]]
SN_summary = SN_summary.rename(columns={"Purchase ID":"Purchase Count"})
SN_summary["Average Purchase Price"] = SN_df.mean()[["Price"]]
SN_summary["Total Purchase Value"] = SN_df.sum()[["Price"]]

# Sort the total purchase value column in descending order
# DF_sorted_name = DF_name.sort_values ("column", ascending=False)  defaults to true
SN_summary = SN_summary.sort_values ("Total Purchase Value", ascending = False)

# Optional: give the displayed data cleaner formatting
SN_summary["Average Purchase Price"] = SN_summary["Average Purchase Price"].map("${:.2f}".format)
SN_summary["Total Purchase Value"] = SN_summary["Total Purchase Value"].map("${:.2f}".format)

# Display a preview of the summary data frame
SN_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
# Retrieve the Item ID, Item Name, and Item Price columns
pop_data = purchase_data [["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
pop_group = pop_data.groupby(["Item ID", "Item Name"])
pop_cnt = pop_group.count()[["Price"]]
pop_price = pop_group.mean()[["Price"]]
pop_total = pop_group.sum()[["Price"]]

# Create a summary data frame to hold the results
pop_summary = pop_group.count()[["Price"]]
pop_summary = pop_summary.rename(columns={"Price":"Purchase Count"})
pop_summary["Item Price"] = pop_price
pop_summary["Total Purchase Value"] = pop_total

# Sort the purchase count column in descending order
pop_summary = pop_summary.sort_values ("Purchase Count", ascending=False) 

# Optional: give the displayed data cleaner formatting
pop_summary["Item Price"] = pop_summary["Item Price"].map("${:.2f}".format)
pop_summary["Total Purchase Value"] = pop_summary["Total Purchase Value"].map("${:.2f}".format)

# Display a preview of the summary data frame
pop_summary.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the data frame

```python
# Create a new df and onvert the formatted columns back to floats
profit_summary = pop_summary
profit_summary[['Item Price']] = profit_summary[['Item Price']].replace('[\$,]','',regex=True).astype(float)
profit_summary[['Total Purchase Value']] = profit_summary[['Total Purchase Value']].replace('[\$,]','',regex=True).astype(float)

# Sort the above table by total purchase value in descending order
profit_summary = profit_summary.sort_values ("Total Purchase Value", ascending=False) 

# Optional: give the displayed data cleaner formatting
profit_summary["Item Price"] = profit_summary["Item Price"].map("${:.2f}".format)
profit_summary["Total Purchase Value"] = profit_summary["Total Purchase Value"].map("${:.2f}".format)

# Display a preview of the data frame
profit_summary.head()
```
